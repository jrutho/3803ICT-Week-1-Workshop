```python
# Write a Python script to download IMDB 250 Top Rated Movies.
# For each movie, you'll need to retrieve the movie title, the ranking, the initial release year, and
# the rating.
# Your data must be stored in a proper imdb_top_250.csv file.

import requests, csv
from bs4 import BeautifulSoup

def get_movies(url):
    response = requests.get(url)
    soup_imdb = BeautifulSoup(response.text)
    print(soup_imdb.prettify())
    all_movies = soup_imdb.find_all("li", {"class":"ipc-metadata-list-summary-item"})

    # using print statements to debug as it is not working properly, this is shown in the cell beneaths output with the 403 forbidden error
    # get movie title :
    titles = [elt.find("a").text for elt in all_movies]
    print(titles)
    # get ranking :
    rankings = [elt.find("span").text for elt in all_movies]
    print(rankings)
    # get release year :
    years = [elt.find("span", {"class":"ipc-metadata-list-item__list-content-item"}).text for elt in all_movies]
    print(years)
    # get rating :
    ratings = [elt.find("span", {"class":"ipc-metadata-list-item__list-content-item"}).text for elt in all_movies]
    print(ratings)
    mega_list = list(zip(titles, rankings, years, ratings))
    mega_list = [list(elt) for elt in mega_list]

    # Write csv
    str_lst = [f"{mega_list[i][1]} / {mega_list[i][0]} ({mega_list[i][2]}) / Starring: {mega_list[i][3]}" \
    for i, val in enumerate(mega_list)]

    res = [elt.split("/") for elt in str_lst]

    with open("imdb_top_250.csv", "w") as f:
        writer = csv.writer(f, delimiter="-")
        writer.writerows(res)
```

```python
def main():
    url = "https://www.imdb.com/chart/top"
    print(f"Downloading {url}")
    get_movies(url)

if __name__ == "__main__":
    main()
```
